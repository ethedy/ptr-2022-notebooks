```
public class MailSender
{
  public void Enviar()
  {
    Console.WriteLine ("Enviando SMTP");
  }

  public string GetEstado()
  {
    return "Todo OK POP3";
  }

  public void Recibir()
  {
    Console.WriteLine ("Recibiendo POP3");
  }
}

public class Messenger
{
  public void Ejecutar()
  {
    MailSender sender = new MailSender();
    
    Console.WriteLine ("Enviando mensajes...");
    sender.Enviar();
    Console.WriteLine ("Recibiendo mensajes...");
    sender.Recibir();
    Console.WriteLine ($"Status: {sender.GetEstado()}");
  }
}

Messenger msg = new Messenger ();

msg.Ejecutar();

==============================================================

//  expression-bodied members
//
public class MailIMAP
{
  public void Enviar() => Console.WriteLine ("Enviando SMTP");
  public string GetEstado() => "Todo OK IMAP";
  public void Recibir() => Console.WriteLine ("Recibiendo IMAP");
  public void Agregado() => Console.WriteLine("No se llama...");
}


==============================================================

public class Messenger
{
  public void Ejecutar()
  {
    MailIMAP sender = new MailIMAP();
    
    Console.WriteLine ("Enviando mensajes...");
    sender.Enviar();
    Console.WriteLine ("Recibiendo mensajes...");
    sender.Recibir();
    Console.WriteLine ($"Status: {sender.GetEstado()}");
  }
}

Messenger msg = new Messenger ();

msg.Ejecutar();



==============================================================

public interface IMailSender
{
  void Enviar();
  void Recibir();
  string GetEstado();
}

public class MailPop3 : IMailSender
{
  public void Enviar() => Console.WriteLine ("Enviando SMTP");
  public string GetEstado() => "Todo OK POP3";
  public void Recibir() => Console.WriteLine ("Recibiendo POP3");
}

public class MailIMAP : IMailSender
{
  public void Enviar() => Console.WriteLine ("Enviando SMTP");
  public string GetEstado() => "Todo OK IMAP";
  public void Recibir() => Console.WriteLine ("Recibiendo IMAP");
}

public class Messenger
{
  private readonly IMailSender _sender;
  
  public Messenger(IMailSender sender)
  {
    _sender = sender;
  }
  
  public void Ejecutar()
  {
    Console.WriteLine ("Enviando mensajes...");
    _sender.Enviar();
    Console.WriteLine ("Recibiendo mensajes...");
    _sender.Recibir();
    Console.WriteLine ($"Status: {_sender.GetEstado()}");
  }
}

  Messenger msg = new Messenger(new MailPop3());
  
  msg.Ejecutar();

  msg = new Messenger(new MailIMAP());
  
  msg.Ejecutar();



Modern Cross-Platform Development Fundamentals
```
